```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
df_main = pd.read_excel('cric_data.xlsx')
```

```python
df_main.head()
```

```python
df_main['start'] = df_main['Span'].apply(lambda x: int(x[0:4]))
```

```python
df_main['end'] = df_main['Span'].apply(lambda x: int(x[5:]))
```

```python
df_main.drop('Span',axis = 1)
```

```python
plt.title("All Batsmen Runs", fontdict={'fontsize': 20, 'fontweight' : 5, 'color' : 'Green'})


df_main.Runs.plot(kind = 'box')
plt.show()
```

```python

plt.figure(figsize = [10,4])
plt.subplot(1,2,1)
plt.title("All Batsmen Average", fontdict={'fontsize': 20, 'fontweight' : 5, 'color' : 'Green'})
df_main.Ave.plot(kind = 'box')
plt.subplot(1,2,2)
plt.title("All Batsmen SR", fontdict={'fontsize': 20, 'fontweight' : 5, 'color' : 'Green'})
df_main.SR.plot(kind = 'box')
plt.show()
```

```python
import seaborn as sns
```

```python
plt.title("Average Sr and run correlation", fontdict={'fontsize': 20, 'fontweight' : 5, 'color' : 'Green'})
sns.heatmap(df_main[["Runs","SR","Ave"]].corr(),annot = True,cmap = 'Greens')
plt.show()
```

```python
df_main.start.value_counts(normalize = True)
```

```python
plt.title("Buckets of avg", fontdict={'fontsize': 20, 'fontweight' : 5, 'color' : 'Green'})
plt.hist(df_main.Ave,bins = 15)
plt.show()
```

```python
plt.title("Buckets of SR", fontdict={'fontsize': 20, 'fontweight' : 5, 'color' : 'Green'})
plt.hist(df_main.SR,bins = 15)
plt.show()
```
